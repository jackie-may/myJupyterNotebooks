```python
import pandas as pd
from scipy import stats
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
premium = diamonds[diamonds['cut'] == 'Premium']
premium.reset_index(inplace=True)

premium.head()
```

```python
good = diamonds[diamonds['cut'] == 'Good']
good.reset_index(inplace=True)

good.head()

#now we have two different dataframes with these two classes
```

```python
#univariate hypothesis test
#both means of carat same in both classes

stats.ttest_ind(premium['carat'], good['carat'])
```

```python
#multivariate analysis
#carat and price correlation

fig = plt.figure(figsize=(16, 6.5))
sns.pairplot(diamonds[['carat','price']])
plt.show()
```
